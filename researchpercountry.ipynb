# Gross domestic spending on R&D: Sweden

This notebook makes simple calculations on OECD data on research spending, describes as follows:

> Gross domestic spending on R&D is defined as the total expenditure (current and capital) on R&D carried out by all resident companies, research institutes, university and government laboratories, etc., in a country. It includes R&D funded from abroad, but excludes domestic funds for R&D performed outside the domestic economy. This indicator is measured in million USD and as percentage of GDP. ([Source: OECD website](https://data.oecd.org/rd/gross-domestic-spending-on-r-d.htm))


```python
%matplotlib inline
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Dataset "Full Indicator Data" downloaded from https://data.oecd.org/rd/gross-domestic-spending-on-r-d.htm  
# (2016-11-27), saved as a csv file. 
df = pd.read_csv('researchpercapita.csv')
```

```python
# Just check out the data. 
df.head()
```

### Create two sub-datasets for country Sweden:

* dfswe = absolute numbers in Million USD.
* dfswegdp = spending as percentage of GDP

```python
sweden = df.loc[df['LOCATION'] == 'SWE']
swedenusd = sweden.loc[sweden['MEASURE'] == 'MLN_USD']
swedenpcgdp = sweden.loc[sweden['MEASURE'] == 'PC_GDP']
dfswe = swedenusd.set_index(swedenusd.TIME)
dfswegdp = swedenpcgdp.set_index(swedenpcgdp.TIME)
```

### Research spending (Million dollars)

```python
f, ax = plt.subplots(figsize=(15, 6))
sns.tsplot([dfswe.Value], time=[dfswe.TIME], ax=ax)
```

### Research spending, percentage of GDP

```python
f, ax = plt.subplots(figsize=(15, 6))
sns.tsplot([dfswegdp.Value], time=[dfswegdp.TIME], ax=ax)
```
